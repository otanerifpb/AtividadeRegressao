# Exercício 01 - Dataset "car_crashes" do seaborn

1) Carregar o dataset "car_crashes"do seaborn

```python
# Importes necessários
import seaborn as sns
from sklearn import datasets, linear_model

# Carregamendo do Dataset
df = sns.load_dataset("car_crashes")

#df.head(10)
```

2) Dropar a coluna "abbrev"

```python
df = df.drop(['abbrev'],axis=1)

df.head(10)
```

3) Separar o dataframe, deixando a coluna 'total' para o target (coluna objetivo - y) e o restante para o X

```python
# Dados para Target(y)
y = df['total'].copy()

# Dados para Features(X)
X = df.drop('total', axis=1).copy()

#y.head(10)
#df.head(10)
#X.head(10)
```

4) Normalizar todo o X

```python
# Importes necessários
from sklearn.preprocessing import MinMaxScaler
import numpy as np

#df['speeding'] = MinMaxScaler().fit_transform(np.array(df['speeding']).reshape(-1,1))

#X['speeding'] = MinMaxScaler().fit_transform(np.array(X['speeding']).reshape(-1,1))
#X['alcohol'] = MinMaxScaler().fit_transform(np.array(X['alcohol']).reshape(-1,1))
#X['not_distracted'] = MinMaxScaler().fit_transform(np.array(X['not_distracted']).reshape(-1,1))
#X['no_previous'] = MinMaxScaler().fit_transform(np.array(X['no_previous']).reshape(-1,1))
#X['ins_premium'] = MinMaxScaler().fit_transform(np.array(X['ins_premium']).reshape(-1,1))
#X['ins_losses'] = MinMaxScaler().fit_transform(np.array(X['ins_losses']).reshape(-1,1))

X[df.columns[1:]] = MinMaxScaler().fit_transform(df[df.columns[1:]])

#df.head(10)
#X.head(10)
```

5) Separar os dados em treinamento e teste

```python
# Importes necessários
from sklearn.model_selection import train_test_split

# Separar dados para Regressão, não possue 'stratify = y'
res = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=None)
X_train, X_test, y_train, y_test = res
```

6) Treinar um modelo linear

```python
# Importes necessários
from sklearn import datasets, linear_model
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

model1 = linear_model.LinearRegression()
model2 = DecisionTreeRegressor()
model3 = KNeighborsRegressor()

treino1 = model1.fit(X_train, y_train)
treino2 = model2.fit(X_train, y_train)
treino3 = model3.fit(X_train, y_train)

predictions1 = treino1.predict(X_test)
predictions2 = treino2.predict(X_test)
predictions3 = treino3.predict(X_test)

print('Regreção Modelo Linear')
print('Resultado: ' + str(mean_squared_error(y_test, predictions1)))
print()
print('Regreção Árvore Decisão')
print('Resultado: ' + str(mean_squared_error(y_test, predictions2)))
print()
print('Regreção K-NN')
print('Resultado: ' + str(mean_squared_error(y_test, predictions3)))
```

# Exercício 02 - Dataset Parkinsons (dataset com 2 classes)

1) Carregar o dataset "parkinsons" pela URL abaixo:

```python
# Importes necessários
import pandas as pd

url = "https://raw.githubusercontent.com/tmoura/machinelearning/master/parkinsons.data"

df1 = pd.read_csv(url, header=None)
columns = len(df1.columns)

#df1.head(10)
```

2) Separar os dados y e X. A coluna y é a coluna 0 (zero)

```python
y = df1[0]

X = df1.loc[:,1:columns-1]

#y.head(10)
#X.head(10)
```

3) Normalizar todas as colunas de X (usando o mesmo código da questão anterior)

```python
# Importes necessários
from sklearn.preprocessing import MinMaxScaler

X[df1.columns[1:]] = MinMaxScaler().fit_transform(df1[df1.columns[1:]])

#df1.head(10)
#X.head(10)
```


4) Separar oa dados em treinamento e teste para Classificação

```python
# Importes necessários
from sklearn.model_selection import train_test_split

# Separar dados para Regressão, possue 'stratify = y'
res = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=None, stratify = y)
train_data, test_data, train_labels, test_labels = res
```

5) Treinar um modelo de regressão logística

```python
# Importes necessários
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import datasets, linear_model
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

model2 = LogisticRegression()

treino2 = model2.fit(train_data, train_labels)

predictions2 = treino2.predict(test_data)

acc2 = metrics.accuracy_score(predictions2, test_labels)
show2 = round(acc2 * 100)

print('Modelo Regressão Logistica para Classificação')
print("Regressão Logistica para Classificação " + "Acurácia: {}%".format(show2))
print(list(predictions2))
print(list(test_labels))
```

6) Treinar uma árvore de decisão

```python
# Importes necessários
from sklearn.tree import DecisionTreeClassifier

model3 = DecisionTreeClassifier()
treino3 = model3.fit(train_data, train_labels)

predictions3 = treino3.predict(test_data)
acc3 = metrics.accuracy_score(predictions3, test_labels)
show3 = round(acc3 * 100)

print('Modelo Classificação Árvore Decisão')
print("Classificação Árvore Decisão " + "Acurácia: {}%".format(show3))
print(list(predictions3))
print(list(test_labels))
```

7) Treinar um KNN para classificação

```python
# Importes necessários
from sklearn.neighbors import KNeighborsClassifier

model4 = KNeighborsClassifier()
treino4 = model4.fit(train_data, train_labels)

predictions4 = treino4.predict(test_data)

acc4 = metrics.accuracy_score(predictions4, test_labels)
show4 = round(acc4 * 100)

print('Modelo Classificação K-NN')
print("Classificação K-NN " + "Acurácia: {}%".format(show4))
print(list(predictions4))
print(list(test_labels))
```

# Exercício 03 - Dataset attention do seaborn

1) Carregar o dataset "attention" do seaborn

```python
# Importes necessários
import seaborn as sns
from seaborn import load_dataset

df2 = sns.load_dataset("attention")

df2.head()
```

2) Fazer um LabelEncoder na coluna "attention"

```python
# Importes necessários
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

df2['attention']= label_encoder.fit_transform(df2['attention'])

df2.head()
```

3) Separar o dataframe, deixando a coluna 'attention' para o target (coluna objetivo - y) e as colunas "solutions" e "score" para o X

```python
y = df2["attention"]

X = df2[["solutions","score"]]

#y.head(10)
X.head()
```

4) Normalizar todo o X usando o mesmo código da questão 1

```python
# Importes necessários
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd

#X['solutions'] = MinMaxScaler().fit_transform(np.array(X['solutions']).reshape(-1,1))
#X['score'] = MinMaxScaler().fit_transform(np.array(X['score']).reshape(-1,1))

dfx = X.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(dfx)
X = pd.DataFrame(x_scaled)

#df2.head(10)
X.head()
```

5) Separar os dados em treinamento e teste

```python
# Importes necessários
from sklearn.model_selection import train_test_split

res = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=None, stratify = y)
X_train, X_test, y_train, y_test = res
```

```python
# Importes necessários
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

model5 = LogisticRegression()
treino5 = model5.fit(X_train, y_train)

predictions5 = treino5.predict(X_test)

acc5 = metrics.accuracy_score(predictions5, y_test)
show5 = "{:.3f}".format(acc5*100)
#show5 = round(acc5 * 100)

print('Modelo Regreção Logistica para Classificação')
print("Regreção Logistica para Classificação " + "Acurácia: {}%".format(show5))
print(list(predictions5))
print(list(y_test))
print(treino5)
```
